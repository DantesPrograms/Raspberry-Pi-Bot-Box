# Raspberry Pi Bot Pins

This notebook reads the repo `.env` file and shows the OLED and joystick pin config in one place.

```python
from pathlib import Path

env_path = Path('.env')
if not env_path.exists():
    raise FileNotFoundError('Missing .env in the repo root')

config = {}
for raw_line in env_path.read_text(encoding='utf-8').splitlines():
    line = raw_line.strip()
    if not line or line.startswith('#') or '=' not in line:
        continue
    key, value = line.split('=', 1)
    config[key.strip()] = value.strip()

config
```

```python
print('OLED')
print(f"  SDA GPIO: {config.get('OLED_SDA_GPIO')}")
print(f"  SCL GPIO: {config.get('OLED_SCL_GPIO')}")
print(f"  SDA physical pin: {config.get('OLED_SDA_PHYSICAL_PIN')}")
print(f"  SCL physical pin: {config.get('OLED_SCL_PHYSICAL_PIN')}")
print(f"  I2C device: {config.get('OLED_I2C_DEVICE')}")
print(f"  I2C address: {config.get('OLED_I2C_ADDRESS')}")
print(f"  Controller: {config.get('OLED_CONTROLLER')}")

print('\nJoystick (legacy notes)')
print(f"  VRX GPIO: {config.get('JOYSTICK_VRX_GPIO')}")
print(f"  VRY GPIO: {config.get('JOYSTICK_VRY_GPIO')}")

print('\nDashboard')
print(f"  Page switch seconds: {config.get('DASHBOARD_PAGE_SECONDS')}")
print(f"  Refresh ms: {config.get('DASHBOARD_REFRESH_MS')}")
print(f"  Bot status file: {config.get('BOT_STATUS_FILE')}")
```

## Notes

- `0x78 / 0x7A` on the OLED module is commonly seen as `0x3C / 0x3D` in Linux tools.
- The current C++ dashboard auto-rotates pages, so the joystick pins are stored here only as reference.
